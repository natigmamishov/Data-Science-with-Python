```python
plt.plot(experience, data_scientists_salary, color='purple',
         label='Salary by experience', marker='o', markeredgecolor='red')

plt.plot(experience, [average_salary]*len(experience), label=f'Average salary-> {average_salary}', color='green')

for i, txt in enumerate(data_scientists_salary):
    plt.annotate(txt, (experience[i], data_scientists_salary[i]),
                 textcoords="offset points", xytext=(0, 10), ha='center', fontsize=8)

plt.text(sum(plt.xlim())/3, average_salary, f'{average_salary:.2f}', color='r', ha='left', va='bottom')

plt.xlim(0, max(experience))
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('String title')
plt.legend()
plt.show()
```

##### Pie chart

Pie charts are circular representations, divided into sectors. The sectors are also called wedges. The arc length of each sector is proportional to the quantity we are describing. It is an effective way to represent information when we are interested mainly in comparing the wedge against the whole pie, instead of wedges against each other.

```python
plt.figure(figsize=(8, 6))
df['Product'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Pie chart of Product Sales")
plt.show()
```

```python
products=df['Product'].value_counts()

print()
```

```python
explode = (0, 0.1, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig, ax = plt.subplots()
ax.pie(products.values, explode=explode, labels=products.index, autopct='%1.1f%%',
       shadow=True, startangle=90)
plt.show()
```

```python
#colors
colors = ['#ff9999','#66b3ff','#99ff99']
 
fig1, ax1 = plt.subplots()
ax1.pie(products.values, colors = colors, labels=products.index, autopct='%1.1f%%', startangle=90)
#draw circle
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.tight_layout()
plt.show()
```

```python
df.groupby("Product")['Gender'].value_counts()
```

```python
product_gender=df.groupby("Product")['Gender'].value_counts()
```

```python
female_counts = product_gender.unstack()['Female'].values

male_counts = product_gender.unstack()['Male'].values
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

#Pie chart
labels = products.index
sizes = products.values
colors = ['blue', 'purple', 'green']

pie = axs[0].pie(sizes, labels=[f'{label}\n{size}%' for label, size in zip(labels, sizes)], 
                 colors=colors, autopct='', startangle=90)
axs[0].set_title('Pie Chart')

#Bar chart
bar_width = 0.35
index = range(len(labels))

bar1 = axs[1].bar(index, female_counts, bar_width, label='Female', color='pink')
bar2 = axs[1].bar([i + bar_width for i in index], male_counts, bar_width, label='Male', color='blue')

axs[1].set_xlabel('Products')
axs[1].set_ylabel('Number of people')
axs[1].set_title('Gender distribution')
axs[1].set_xticks([i + bar_width/2 for i in index])
axs[1].set_xticklabels(labels)
axs[1].legend()

# Writing numbers on the bars
for i, value in enumerate(female_counts):
    axs[1].text(i, value + 0.1, str(value), ha='center', va='bottom', color='black', fontweight='bold')

for i, value in enumerate(male_counts):
    axs[1].text(i + bar_width, value + 0.1, str(value), ha='center', va='bottom', color='black', fontweight='bold')

plt.tight_layout()
plt.show()
```

##### Boxplot

Boxplot allows us to compare distributions of values by showing the median, quartiles, maximum and minimum of a set of values.

```python
plt.boxplot(x='Age', data=df)
plt.ylabel('Value')
plt.title("Simple Box Plot")
plt.show()
```

```python
plt.boxplot(x='Age', data=df, vert=False)
plt.ylabel('Value')
plt.title("Simple Box Plot")
plt.show()
```

```python
plot = plt.boxplot(df['Age'], patch_artist=True, showmeans=True)

edge_color = 'orange'
fill_color = 'lightgreen'
marker_color = 'red'

for prop in ['boxes', 'whiskers', 'means', 'medians', 'caps']:
    plt.setp(plot[prop], color=edge_color, linewidth=1.5)

for prop in ['boxes']:
    plt.setp(plot[prop], facecolor=fill_color) 

for prop in ['fliers', 'means']:
    plt.setp(plot[prop], markerfacecolor=marker_color, 
             markeredgecolor=marker_color)

plt.show()
```
