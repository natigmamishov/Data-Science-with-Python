### Hypothesis Testing

Hypothesis testing is an act in statistics with which an analyst tests an assumption regarding a population parameter. It is used extensively to assess the plausibility of a hypothesis by using sample data. The assumptions made might be correct or it may not be. Data alone isn't interesting enough. It is the interpretation that we are really interested in

In research, there are two types of hypotheses: null and alternative. They work as a complementary pair, each stating that the other is wrong.

**Null Hypothesis (H0)** ‚Äì> This can be thought of as the implied hypothesis. ‚ÄúNull‚Äù meaning ‚Äúnothing.‚Äù  This hypothesis states that there is no difference between groups or no relationship between variables. The null hypothesis is a presumption of status quo or no change.

**Alternative Hypothesis (H1)** ‚Äì> This is also known as the claim. This hypothesis should state what you expect the data to show, based on your research on the topic. This is your answer to your research question.

Example: Testing the Mean of a Population

Scenario: A manufacturer claims that the average lifespan of their light bulbs is 1000 hours. A quality control engineer wants to test if this claim is accurate.

Hypotheses:

‚Ä¢ **Null Hypothesis (ùêª0)**: The mean lifespan of the light bulbs is 1000 hours (Œº=1000).          
‚Ä¢ **Alternative Hypothesis (ùêª1)**: The mean lifespan of the light bulbs is not 1000 hours (ùúá‚â†1000).

Hypothesis testing can be set up in 3 different ways:

1. **Two-Tailed Test**         
‚Ä¢ Null Hypothesis (ùêª0): The sample mean is equal to a specified population mean.         
‚Ä¢ Alternative Hypothesis (H1): The sample mean is not equal to the specified population mean.
$$ H0: Œº=Œº_{0} $$
$$ H1: ùúá‚â†ùúá_{0} $$                        

Note: This test checks if the sample mean differs significantly from the specified population mean in either direction.
                             

2. **Right-Tailed Test**               
‚Ä¢ Null Hypothesis (H0): The sample mean is less than or equal to a specified population mean.       
‚Ä¢ Alternative Hypothesis (H1): The sample mean is greater than the specified population mean.
$$ H0: ùúá‚â§ùúá_{0}  $$ 
$$ H1:ùúá>ùúá_{0}   $$
 
Note: This test determines if the sample mean is significantly greater than the specified population mean.

3. **Left-Tailed Test**           
Null Hypothesis (H0): The sample mean is greater than or equal to a specified population mean.       
Alternative Hypothesis (H1): The sample mean is less than the specified population mean.
$$ H0: ùúá‚â•ùúá_{0}    $$
$$ H1: ùúá<ùúá_{0}    $$

**Choose the Alpha (Œ±) Level:**

‚Ä¢ The alpha level sets the acceptable risk of making a Type I error (rejecting the null hypothesis when it is actually true). It is typically set at 0.05 or 0.01.

**Determine the Test Statistic:**

‚Ä¢ Select the appropriate statistical test (e.g., t-test, ANOVA, chi-square test, etc.) that matches your data and hypothesis.
This test statistic provides a mathematical formula to evaluate your data

**Calculate the Test Statistic and Find the p-value:**

‚Ä¢ Compute the test statistic based on your sample data.
Use this statistic to calculate the p-value (the probability of observing your data if the null hypothesis is true).

**Make a Decision:**

‚Ä¢ Compare the p-value to the alpha level:            
    ‚Ä¢ If p-value ‚â§ Œ±: Reject the null hypothesis. The data provide sufficient evidence to support the alternative hypothesis.           
    ‚Ä¢ If p-value > Œ±: Fail to reject the null hypothesis. The data do not provide sufficient evidence to support the alternative hypothesis.

**Interpret the Results:**           

‚Ä¢ If you reject the null hypothesis, interpret what the results mean in terms of practical significance or implications.           
‚Ä¢ If you fail to reject the null hypothesis, conclude that the data do not show a statistically significant difference or relationship.

**Report the Findings:**            

‚Ä¢ Present your findings along with the details of the statistical test used.          
‚Ä¢ Explain the data analysis and results in a way that supports the accuracy and reliability of your findings.

### Test for normality

Before running any hypothesis test, it's important to know which statistical method we should use. Statstical methods are divided in two parts :

‚Ä¢ Parametric statistical methods

‚Ä¢ Nonparametric statistical methods

To know which one to use, normality tests must be done on our data. If our data have a known and specific distribution, such as the Gaussian distribution; parametric statistical methods must be used. On the contrary, if data are not Gaussian, nonparametric statistical methods should be used.

There are two main ways to know if our data are Gaussian :

1. Graphical methods

‚Ä¢ Stem and leaf display    
‚Ä¢ Histogram plot                
‚Ä¢ QQ plot

2. Statistical methods

‚Ä¢ Shapiro test       
‚Ä¢ D'Agostino and Pearson test       
‚Ä¢ Anderson-Darling test          
‚Ä¢ Kolmogorov-Smirnov test       

**Graphical methods** are mainly used for qualifying deviations from normality

### One sample t-test

**If the data we have is greater than 120 or a normal distribution, we can apply the t test.**

**Example:** According to the American Health Association, a pregnant woman‚Äôs average blood pressure is 120 mm Hg. Collected 15 random samples from pregnant women to check whether the sample blood pressure differs from accepted standard blood pressure. Significance level:0.05       
[131, 115, 118, 120, 126, 137, 125, 116, 117, 122, 123, 128, 110, 124, 133]

‚Ä¢  Null Hypothesis: H0: Œº=120         
‚Ä¢  Alternative Hypothesis: H1: Œº‚â†120              
‚Ä¢  Œ±=0.05                                                
‚Ä¢  Degrees of freedom(df): 15 ‚Äì 1 = 14                   

Calculate the critical value

**$$ t_{critic}=2.145 $$**

If the calculated t value is less than -2.145 or greater than 2.145, then reject the Null Hypothesis.

**Test Statistics**              
xÃÖ = 123                        
Œº0 = 120           
std=7.3

$$ t_{statistics}=1.59 $$

```python
example=[131, 115, 118, 120, 126, 137, 125, 116, 117, 122, 123, 128, 110, 124, 133]

mean_value = np.mean(example)

print("Mean:",mean_value)

t_statistic, p_value=stats.ttest_1samp(example, 120, alternative='two-sided')

t_c=stats.t.ppf(q=1-.05/2,df=len(data)-1)

print(f't critic: {t_c}')

print("P Value:", p_value)

print("t-test Value:", t_statistic)

# 0.05 or 5% is significance level or alpha.

if p_value < 0.05:    

    print("Hypothesis rejected")

else:
    
    print("Hypothesis fail to rejected")
```

**Interpret the results:** Compare t calc to t critical. In hypothesis testing, a critical value is a point on the test distribution compared to the test statistic to determine whether to reject the Null Hypothesis. The calculated t-statistic value is less than the critical value; hence we failed to reject Null Hypothesis.

The number of data is less than 120, but since it is a normal distribution, we can apply the t test.

```python
t_statistic, p_value=stats.ttest_1samp(data, 60, alternative='two-sided')
print(f't statistic: {t_statistic}')

t_c=stats.t.ppf(q=1-.05/2,df=len(data)-1)
print(f't critic: {t_c}')
```

Since t_statictic value is less than t_c value, t_statictic belongs to a rejection region. In this case, H0 is rejected.

Since we reject H0, a is either greater than 60 or less.

H0 (null hypothesis): mu>=60                                                                           
H1 (alternative hypotesis): mu<60

```python
t_statistic, p_value=stats.ttest_1samp(data, 60, alternative='less')
print(f't statistic: {t_statistic}')

t_c=stats.t.ppf(q=.05,df=len(data)-1)
print(f't critic: {t_c}')
```

Since t_statictic value is less than t_c value, t_statictic belongs to a rejection region. In this case, H0 is rejected. Since a is rejected, it appears that the mean is less than 60

### Confidence Interval

**A confidence interval** is a range of values, derived from sample data, that is likely to contain the true value of a population parameter (such as a mean or proportion) with a certain level of confidence. It provides an estimate of how much uncertainty is associated with the sample statistic.

Example:
If a researcher calculates a 95% confidence interval for the average height of a population and finds it to be [170cm,180cm], this means that there is a 95% chance that the true population mean height lies within that range.

```python
confidence_interval = stats.t.interval(
    confidence=0.95,        
    df=len(data)-1,      
    loc=np.mean(data),
    scale=stats.sem(data)
)

print(confidence_interval)
```

### Wilcoxon test

H0 (null hypothesis): mu=25                                            
H1 (alternative hypotesis): mu!=25

```python
w, p = stats.wilcoxon(np.array(data)-25, alternative='two-sided')
print(w, p)
```

H0 is rejected because p value is less than 0.05. Then the median is either greater than 25 or less. We set up a new hypothesis test.

H0 (null hypothesis): mu=25                                            
H1 (alternative hypotesis): mu<25

```python
w, p = stats.wilcoxon(np.array(data)-25, alternative='less')
print(w, p)
```

H0 is rejected because p value is less than 0.05. As can be understood then, the median is considered to be less than 25.
