### Seaborn

**Seaborn is a statistical plotting library that is specifically designed to interact well with Pandas DataFrames to create common statistical plot types**

**Seaborn is built directly off of Matplotlib but uses a simpler "one-line" syntax**

**Common student question: How do A choose which plot to use?**

```python
sns.scatterplot(x='salary', y='sales', data=df)
plt.show()
```

```python
sns.scatterplot(x='salary', y='sales', hue='level of education', data=df)
plt.show()
```

```python
sns.scatterplot(x='salary', y='sales', hue='level of education', style='level of education', data=df)
plt.show()
```

```python
sns.scatterplot(x='salary', y='sales', hue='level of education', data=df, palette='Dark2')
plt.show()
```

```python
sns.scatterplot(x='salary', y='sales', hue='level of education', data=df, palette='Dark2',
               size='salary', alpha=0.2)
plt.show()
```

```python
sns.rugplot(x='salary', data=df)
plt.show()
```

```python
sns.histplot(x='salary', data=df)
plt.show()
```

```python
sns.histplot(x='salary', data=df, bins=20, color='purple', edgecolor='blue', kde=True)
plt.show()
```

```python
sns.countplot(x='level of education', data=df)
plt.show()
```

```python
sns.countplot(x='level of education', data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(x='level of education', data=df, hue='division', palette='Set2')
plt.xticks(rotation=90)
plt.show()
```

```python
#barplot showing mean and sd bar
sns.barplot(x='level of education', y='salary', data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.boxplot(y='math score', data=df)
plt.show()
```

```python
sns.boxplot(y='math score', x='test preparation course',data=df)
plt.show()
```

```python
sns.boxplot(y='math score', x='parental level of education', hue='test preparation course',
            palette='Set2', data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.violinplot(x='reading score', y='parental level of education', data=df)
plt.show()
```

```python
sns.violinplot(x='reading score', y='parental level of education', data=df, inner='quartile')
plt.show()
```

```python
sns.swarmplot(x='math score', size=2, y='gender', hue='test preparation course', data=df)
plt.show()
```

```python
sns.boxenplot(x='math score', data=df)
plt.show()
```

```python
sns.boxenplot(x='math score', y='test preparation course', data=df)
plt.show()
```

```python
sns.boxenplot(x='math score', y='test preparation course', hue='gender', data=df)
plt.show()
```

```python
sns.jointplot(x='math score', y='reading score', kind='hex', data=df)
plt.show()
```

```python
sns.jointplot(x='math score', y='reading score', kind='scatter', data=df)
plt.show()
```

```python
sns.jointplot(x='math score', y='reading score', kind='scatter', hue="gender", data=df)
plt.show()
```

```python
sns.lmplot(x='reading score', y='writing score', hue='gender', col='gender', data=df)
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
sns.pairplot(df, hue='gender')
plt.show()
```

```python
sns.pairplot(df, hue='gender', diag_kind='hist')
plt.show()
```

```python
df.columns
```

```python
sns.catplot(x='gender', y='math score', kind='box', hue='lunch', row='test preparation course', data=df)
plt.show()
```

```python
g=sns.PairGrid(df, hue='gender')
g=g.map_upper(sns.scatterplot)
g=g.map_diag(sns.histplot)
g=g.map_lower(sns.kdeplot)
g=g.add_legend()
plt.show()
```

```python
num_cols=df._get_numeric_data().columns

print(num_cols)
```

```python
df[num_cols].corr()
```

```python
corr=df[num_cols].corr()
```

```python
sns.heatmap(corr)
plt.show()
```

```python
sns.heatmap(corr, annot=True, cmap='viridis')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.title('Correlation Analysis',color='Red',fontsize=20,pad=40)
mask = np.triu(np.ones_like(corr, dtype = bool))
sns.heatmap(corr, mask=mask, annot=True, linewidths=.5);
plt.xticks(rotation=60)
plt.yticks(rotation = 60)
plt.show()
```
