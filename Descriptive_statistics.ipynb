## Descriptive statistics

Descriptive statistics gives us the basic summary measures about the dataset. The summary measures include measures of central tendency (mean, median and mode) and measures of variability (variance, standard deviation, minimum/maximum values, IQR (Interquartile Range), skewness and kurtosis). I have used the fortune 500 dataset from the data world website for this project.

### Populations and Samples

Population-> Includes all of the elements from a set of data.              
Sample-> Consists of one or more observations from the population.            
Observation unit->The smallest individual unit observed or measured during data collection.

### Variable

• **dependent variable**          
• **Independent variable**

The independent variable is the variable that you change or control in an experiment or study. This variable is used to observe its effect on other variables. In other words, the change in the independent variable allows us to observe how it affects the dependent variable.

• In an agricultural study, the types of fertilizers given to plants (organic fertilizer, chemical fertilizer) are independent variables.     
• Observing whether the plant grows faster.

### Parameter and statistic

A **parameter** is a numerical value that describes a characteristic of a population. It is usually denoted by Greek letters (e.g., μ for population mean).                       
Example: The average height of all adults in a country.

A **statistic** is a numerical value that describes a characteristic of a sample, which is a subset of the population. It is usually denoted by Latin letters (e.g., x̄ for sample mean).

Example: The average height of 100 adults selected from the population.

Key Difference:                                
Parameter: Describes the entire population.              
Statistic: Describes a sample and is used to estimate the population parameter.

### Central tendency

The following definitions of mean, median, and mode state in words how each is calculated.

• **Mean:** The mean of a data set is the average value. Is that circular? Let me state this in the form of an algorithm. To find the mean of a data set first sum up all of the values in the data set then divide by the number of data points or examples in the data set. When we find or know the mean of an entire population we call that mean a parameter of the population and in is assigned the symbol μ. When the mean is of a sample it is a statistic of the sample ans is assigned the symbol x̄.

• **Median:** The median of a data set is simply the data point that occurs exactly in the middle of the data set. To find the median first the data needs to be sorted and listed in ascending or descending order. Then if there is an odd number of data points in the set the median is in the exact middle of this list. If there is an even number of data points in the list then the median is the average of the two middle values in the list.

• **Mode:** The mode is the value in a data set that is repeated the most often. Below the mean, median, and mode are calculated for a dummy data set.

```python
# Define the data set
data=[13, 10, 15, 12, 17, 12, 19, 18, 11, 12, 190]
```

If the data series we have is symmetric, the arithmetic mean is the best measure of central tendency. Otherwise, the median should be considered.

```python
print(np.mean(data))
print(np.median(data))
print(st.mode(data)[0])
```

• The arithmetic mean is much higher than the median. This indicates the presence of some very large values in the data set that are pulling the mean upwards.                    
• The median represents the middle value of the data set and is not affected by extreme values.           
In such data sets, the median is a more appropriate measure of central tendency.

##### Measures of Dispersion

• **Variance:** The average of the squared distance of the mean. Variance is interesting in a mathematical sense, but the standard deviation is often a much better measure of how spread out the distribution is.

• **Standard Deviation:** Once the variance is in hand, standard deviation is eacy to find. It is simply the square root of the variance. The symbol for population standard deviation is σ while the simple for sample standard deviation is s or sd.                      

• **Range:** Range is the diffrence between the highest and the lowest value of data. Range is affected by the outliers (data points are the value much higher or lower than the rest of the data). $$ range = x_{max} - x_{min}  $$

```python
y = np.random.uniform(5,8,100)
x1 = np.random.uniform(10,20,100)
x2 = np.random.uniform(0,30,100)
```

```python
plt.scatter(x1,y,color="black")
plt.scatter(x2,y,color="orange")
plt.xlim([-1,31])
plt.ylim([2,11])
plt.xlabel("x")
plt.ylabel("y")
print("X1 mean: {} and meadian: {}".format(np.mean(x1),np.median(x1)))
print("X2 mean: {} and meadian: {}".format(np.mean(x2),np.median(x2)))
```

Having the same arithmetic mean for two data sets does not imply that their distributions or structures are the same. Especially when you mentioned that visually they appear different, the reason could be differences in their measures of variability.

```python
print(np.var(x1))
print(np.var(x2))
```

What can we infer from these results?

1. Magnitude of Variance:

• For var1, the variance of 9.56 is relatively small, indicating that the data points in var1 are less spread out around the mean.        
• For var2, the variance of 72.46 is much larger, indicating that the data points in var2 are more spread out and more variable.

2. Standard Deviation:

• Standard deviation, which is the square root of variance, provides a more intuitive measure of the spread of data around the mean. Calculating the standard deviations for var1 and var2 would further clarify these values.

```python
print(np.std(x1))
print(np.std(x2))
```

• For std1, the standard deviation is 3.09, indicating that the data points in std1 are relatively close to the mean, with less variability compared to std2.          
• For std2, the standard deviation is 8.51, which is significantly larger than std1, indicating that the data points in std2 are more spread out and more variable around the mean.

##### Measures of Position

• **Quartiles:** Quartiles are statistical values that divide a data set into four equal parts. They are used to better understand the distribution of the data and provide insights into central tendency and spread. There are four quartiles:
1. Q1: Shows the distribution of data in the lower quartile.
2. Q2 (Median): Represents the midpoint of the data set.
3. Q3: Shows the distribution of data in the upper quartile.
4. Q1 and Q3: The difference between these two values, known as the Interquartile Range (IQR), indicates the spread of the middle 50% of the data set.

```python
# Calculate quartiles
Q1 = np.percentile(data, 25)
Q2 = np.percentile(data, 50)  # Median
Q3 = np.percentile(data, 75)

# Print the results
print(f"Q1 (First Quartile): {Q1}")
print(f"Q2 (Median): {Q2}")
print(f"Q3 (Third Quartile): {Q3}")
```

```python
# Identify outliers using the IQR method
Q1 = df['Years'].quantile(0.25)
Q3 = df['Years'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Years'] < lower_bound) | (df['Years'] > upper_bound)]
print("Outliers:")
print(outliers)
```

```python
# Visualize the data
sns.boxplot(x=df['Years'])
plt.title('Box Plot with Outliers')
plt.show()
```

```python
def get_outliers(dataframe, column):
    
    Q1=dataframe[column].quantile(0.25)
    Q3=dataframe[column].quantile(0.75)
    IQR=Q3-Q1
    
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    
    return dataframe[ (dataframe[column]<lower_bound) | (dataframe[column]>upper_bound) ].index
```

```python
for col in df._get_numeric_data().columns:
    
    print(col, '->', get_outliers(df, col))
```

##### Skewness

Skewness is a statistical term that measures the asymmetry of a probability distribution or a dataset. It describes the degree of distortion from the symmetrical bell curve (normal distribution).

The value of Pearson Skewness indicates:

• Close to 0: The dataset is relatively symmetric.           
• Negative: The dataset is negatively skewed (left-skewed, where the left tail is longer).         
• Positive: The dataset is positively skewed (right-skewed, where the right tail is longer).

```python
# Plotting histogram and boxplot
plt.figure(figsize=(10, 5))

# Histogram plot
plt.hist(df['Years'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histogram of Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Calculate Pearson Skewness
mean = np.mean(df['Years'])
median = np.median(df['Years'])
std_dev = np.std(df['Years'])

pearson_skewness = 3 * (mean - median) / std_dev

print("Pearson Skewness:", pearson_skewness)
```

The Pearson Skewness value of 0.88 indicates that the dataset is positively skewed. Here's how we interpret this:

• Positive Skewness: A Pearson Skewness value greater than 0 suggests that the dataset has a longer right tail. This means that there are more values on the right side of the distribution (higher values) compared to the left side.     
• The magnitude of the skewness (1.0004402900985772 in this case) gives an indication of the degree of asymmetry:    
• A value closer to 0 indicates less skewness and a more symmetrical distribution.      
• A larger positive value indicates stronger positive skewness, meaning the distribution is more heavily skewed towards higher values.

##### Kurtosis

Kurtosis is a statistical measure that describes the degree of peakedness or flatness of a probability distribution compared to a normal distribution. It quantifies the tails and peak of a distribution relative to the normal distribution.

Types of Kurtosis
1. Leptokurtic (Positive Kurtosis):

• A leptokurtic distribution has a sharper peak and heavier tails compared to a normal distribution. This means that data points are more concentrated around the mean, with heavier tails indicating more extreme values.

2. Platykurtic (Negative Kurtosis):

• A platykurtic distribution has a flatter peak and lighter tails compared to a normal distribution. This indicates that data points are more spread out from the mean, with fewer extreme values.

3. Mesokurtic (Normal Kurtosis):

• A mesokurtic distribution closely resembles a normal distribution in terms of its peakedness and tail behavior.

• kurtosis=3 -> Normal Kurtosis             
• kurtosis>3 -> Leptokurtic (Positive Kurtosis)             
• kurtosis<3 -> Platykurtic (Negative Kurtosis)
