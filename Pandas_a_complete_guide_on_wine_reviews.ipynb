##### concatenation and merge

```python
# creating the Series
series1 = pd.Series([12, 3, 27])

series2 = pd.Series(['U', 'Q', 'N'])
```

```python
pd.concat([series1, series2], axis = 0)
```

```python
pd.concat([series1, series2], axis = 1)
```

```python
pd.concat([series1, series2], axis = 1, keys=['col1', 'col2'])
```

```python
df1=pd.DataFrame({'number': [12, 24, 27, 10, 4],
                 'name':['Ul', 'En', 'Sr', 'Gun', 'Ay']
                 })

df2=pd.DataFrame({'number':[12, 10, 24, 27, 10, 12, 3],
                 'subject':['SQL', 'Python', 'Algebra', 'SQL', 'SQL', 'Python', 'C'],
                  'grade':[100, 90, 12, 45, 65, 95, 85]
                 })
```

```python
df1
```

```python
df2
```

```python
df1.merge(df2, how='inner', on='number')
```

```python
df1.merge(df2, how='left', on='number')
```

```python
df1.merge(df2, how='outer', on='number')
```

```python
df1.merge(df2, left_on='number', right_on='grade')
```
